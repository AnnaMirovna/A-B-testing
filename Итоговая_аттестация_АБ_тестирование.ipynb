# Задача 1.
Подведите результаты эксперимента в экселе по следующим данным: ab_stats.csv - Google Диск:
Стат значимо ли отличается ARPPU в двух группах ? Какие рекомендации дадите
менеджеру?

```

!pip install pandas scipy

import pandas as pd
from scipy.stats import ttest_ind
from google.colab import files

data = pd.read_csv('ab_stats(1).csv')

# Шаг 1: Фильтрация пользователей, которые совершили покупки (num_purchases > 0)
data_paying_users = data[data['num_purchases'] > 0].copy()

# Шаг 2: Расчет ARPPU (Average Revenue Per Paying User)
data_paying_users['ARPPU'] = data_paying_users['revenue'] / data_paying_users['num_purchases']

# Шаг 3: Вычисление среднего значения ARPPU для каждой группы
arppu_grouped = data_paying_users.groupby('ab_group')['ARPPU'].mean()

# Шаг 4: Проведение t-теста
arppu_A = data_paying_users[data_paying_users['ab_group'] == 'A']['ARPPU']
arppu_B = data_paying_users[data_paying_users['ab_group'] == 'B']['ARPPU']

# t-тест
t_stat, p_value = ttest_ind(arppu_A, arppu_B)

# Шаг 5: Вывод результатов
print("Среднее значение ARPPU по группам:")
print(arppu_grouped)
print("\nРезультаты t-теста:")
print(f"t-статистика: {t_stat}")
print(f"p-значение: {p_value}")
```

Средние значения ARPPU (средняя выручка на одного покупателя) в двух группах:

  Группа A: 6.55
  Группа B: 6.56

Результаты t-теста:

  t-статистика: -0.013
  p-значение: 0.99

Вывод:

Различие в ARPPU между группами A и B статистически незначимо (p-значение > 0.05). Это означает, что нет достаточных оснований утверждать, что одна из групп показывает лучшее финансовое поведение по сравнению с другой.

Рекомендации для менеджера:
Поскольку различия между группами незначительны, текущие изменения, вероятно, не оказывают влияния на ключевой показатель ARPPU.
Возможно, стоит пересмотреть экспериментальные условия или попробовать другие метрики (например, LTV, конверсию), чтобы выявить значимые различия.
Если важны незначительные улучшения, можно продолжить тестирование, увеличив выборку, чтобы убедиться, что результаты устойчивы.

# Задача 4.
Вы решили сравнивать метрику CPA в двух группах. Размер выборки - 2350 элементов в
каждой группе.
Для проверки нормальности распределения на выборке в 2350 наблюдений применили ,
критерий Шапиро-Уилка и получили p-value, равный 0.00002, alpha = 5%.
Какой бы вывод мы могли сделать в данном случае?
В этом случае какой статистический критерий для проверки первоначальной гипотезы тут лучше
всего подойдёт и почему?

1. Вывод по критерию Шапиро-Уилка

Критерий Шапиро-Уилка используется для проверки нормальности распределения данных. Он возвращает p-value, и интерпретация его заключается в следующем:

  Нулевая гипотеза (H0): Данные нормально распределены.
  Альтернативная гипотеза (H1): Данные не нормально распределены.

Если p-value меньше уровня значимости (α = 5% или 0.05), то мы отвергаем нулевую гипотезу и считаем, что данные не нормально распределены.

В этом случае:

  p-value = 0.00002 — это значительно меньше 0.05.
  Следовательно, мы отвергаем нулевую гипотезу и делаем вывод, что данные не распределены нормально.

2. Выбор статистического критерия

Поскольку данные не являются нормально распределёнными, лучше использовать непараметрический критерий, который не требует нормальности данных. Подходящие методы для сравнения CPA (стоимость привлечения клиента) в двух группах:
Манна-Уитни (U-критерий)

Это непараметрический тест, который используется для сравнения двух независимых групп, когда данные не являются нормально распределёнными.
Он основан на ранговых данных и проверяет, различаются ли распределения двух групп.
Применим для ситуаций, когда данные не следуют нормальному распределению.

Почему нельзя использовать t-тест?

t-тест для независимых выборок предполагает нормальность распределения данных в каждой группе. В нашем случае это условие не выполнено, поэтому t-тест может дать неверные результаты.

Рекомендация:

Используйте U-критерий Манна-Уитни для проверки гипотезы о различии CPA между группами, так как он подходит для анализа данных с нарушением нормальности распределения.

Для того чтобы ответить на вопрос о статистической значимости результатов A/B теста, нам нужно выполнить следующие шаги:
Шаг 1: Определение гипотез

Нулевая гипотеза (H₀): Среднее время, проведённое пользователем на сайте (average timespent per user), в группах A и B не отличается.
Альтернативная гипотеза (H₁): Среднее время, проведённое пользователем на сайте, в группах различается.

Шаг 2: Использование t-теста для двух независимых выборок

Поскольку у нас есть средние, стандартные отклонения и размеры выборок, мы можем использовать t-тест для независимых выборок, чтобы определить, есть ли статистически значимые различия между группами A и B.

Формула t-статистики:

t = XA − XB / SA2nA+SB2nB
t=nA​SA2​​+nB​SB2​​
​XA​ˉ​−XB​ˉ​​

Где:

XAˉXA​ˉ​ и XBˉXB​ˉ​ — средние значения для групп A и B.
SASA​ и SBSB​ — стандартные отклонения для групп A и B.
nAnA​ и nBnB​ — размеры выборок в группах A и B.

Шаг 3: Определение уровня доверия и p-value

Уровень доверия 80% означает, что уровень значимости (α) составляет 20% или 0.20.

Теперь рассчитаем t-статистику и p-value, чтобы определить, есть ли различия.
